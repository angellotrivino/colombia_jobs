### Process data

#### Read data

```python
input_colombia_jobs_file = path.data_raw_dir("conjunto-mercado-laboral-bogota-31012024.xlsx")
input_colombia_jobs_file
```

```python
colombia_jobs_df = pd.read_excel(input_colombia_jobs_file, sheet_name="Mercado laboral Bogotá")
colombia_jobs_df.info()
```

```python
colombia_jobs_df.head()
```

```python
len(colombia_jobs_df)
```

#### Cast columns to typedata

```python
colombia_jobs_df[colombia_jobs_df['Periodo'].str.len() > 4].head()
```

```python
colombia_jobs_df['Periodo'].unique()
```

##### Add New Column Year -> Anio

```python
colombia_jobs_df.insert(0, 'Anio', colombia_jobs_df['Periodo']
                        .astype(str).str[:4].astype('Int32'))
colombia_jobs_df.head()
```

##### Add New Column 'Fecha'

```python
colombia_jobs_df.insert(0, 'Fecha', (colombia_jobs_df['Anio'].astype(str) + '-' + colombia_jobs_df['Trimestre móvil cronológico'].astype(str) + '-1').astype('datetime64[us]'))
colombia_jobs_df.head()
```

##### Convert to Float64 Columns: Tasa de Infromalidad y Totales Informales

```python
condicion = colombia_jobs_df['Tasa  de informalidad DANE '] == '-'
colombia_jobs_df.loc[condicion, 'Tasa  de informalidad DANE '] = 0
colombia_jobs_df['Tasa  de informalidad DANE '] = colombia_jobs_df['Tasa  de informalidad DANE '].astype('Float64')
```

```python
condicion_2 = colombia_jobs_df['Total informales DANE '] == '-'
colombia_jobs_df.loc[condicion_2, 'Total informales DANE '] = 0
colombia_jobs_df['Total informales DANE '] = colombia_jobs_df['Total informales DANE '].astype('Float64')
```

```python
col_jobs_final_df = colombia_jobs_df
```

##### Save output data

```python
output_col_jobs_file = path.data_processed_dir("conjunto-mercado-laboral-bogota-31012024_processed.csv")
```

```python
col_jobs_final_df.to_csv(output_col_jobs_file, index=False)
```
