##### Covert column Fecha to DateTime

```python
col_jobs_df['Fecha'] = col_jobs_df['Fecha'].astype('datetime64[us]')
```

```python
sns.lineplot(
    data=col_jobs_df,
    x="Fecha",
    y="Poblacion en edad de trabajar"
)

plt.xticks(rotation=15)
plt.xlabel("Date: Year")
plt.ylabel("Work Age People")
plt.title("Status Colombia's Jobs");
```

```python
sns.lineplot(
    data=col_jobs_df,
    x="Fecha",
    y="Tasa de desocupacion",
)

plt.xticks(rotation=15)
plt.xlabel("Date: Year")
plt.ylabel("Unemployment rate %")
plt.title("Status Colombia's Jobs");
```

```python
items = ['Anio', 'Poblacion total GEIH DANE', 'Poblacion en edad de trabajar', 'Fuerza de trabajo', 'Ocupados', 'Desocupados']
some_columns_col_jobs_df = col_jobs_df[items]
```

```python
some_columns_col_jobs_df.info()
```

```python
processed_df = (
    some_columns_col_jobs_df
    .pivot_longer(
        index="Anio",
        names_to="items"
    )
    .groupby(["Anio", "items"])
    .sum("value")
)
```

```python
processed_df.all
```

```python
sns.lineplot(
    data=processed_df,
    x="Anio",
    y="value",
    hue="items"
)

plt.xticks(rotation=15)
plt.xlabel("Date: year")
plt.ylabel("Value")
plt.title("Status Colombia's Jobs");
```
